```python
print('importing data_cleaner modules...')

import pandas as pd
import numpy as np
from collections import defaultdict
import re

print('done')
```

```python
def clean_GSE_series_matrix(file): 
    print('reading file...')

    # file='GSE63990_series_matrix.txt'
    with open(file, "r") as fh: dat = fh.read()
    dat = dat.split("\n")

    d = defaultdict(list)                
    for row in dat:                     
        row = row.split("\t")
        col = row[0][1:] # Drop that annoying ! and " at the beginning
        for val in row[1:]: 
            d[col].append(val)

    infectionRegex = re.compile(r'infection_status') # this is the line for the "answers" that the NN needs
    for k in list(d.keys()): # This has to be a list, otherwise it whines about dict size changing during loop
        if len(d[k]) != 280: # Seems to be the number of samples, if not =
            mo = infectionRegex.search(str(d[k])) # then search to see if it's the poorly formatted "answer" line
            if mo is None: 
                print(f"Dropping {k}")
                del d[k]
            else: # if it is,
                print(f"Converting {k}")
                statusRegex = re.compile('bacterial|non-infectious illness|viral') # look for the "answer"
                newmo = statusRegex.findall(str(d[k])) # list of answers
                d[k] = newmo # add to dict

    df = pd.DataFrame.from_dict(d)
    
    return df

    print('done')
```

```python
# returns list of keys which have numeric values
def get_numeric_keys(d): return [k for k in d.keys() if any([i in str(k) for i in '0123456789'])]
```
